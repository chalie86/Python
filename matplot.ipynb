```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'y')
plt.plot(y,x,'p')
```

```python
fig = plt.figure()
axes =  fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.15,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(x,y)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
#plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
    
    
    axes[0].plot (x,y)
    axes[0].set_title('First Plot')
    
    axes[1].plot (y,x)
    axes[1].set_title('Second Plot')
    
    plt.tight_layout()
```

```python
#fig = plt.figure(figsize = (3,2),dpi=100)

#ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y)

fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```
